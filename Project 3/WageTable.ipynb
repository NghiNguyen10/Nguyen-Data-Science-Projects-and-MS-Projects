```python
import numpy as np
import pandas as pd
from os import getcwd, listdir

data_path = getcwd() + "/data/sas_to_pandas/raw/wages"
grow_path = getcwd() + "/data/sas_to_pandas/joins/OCC_List.csv"
df = pd.read_csv(f"{data_path}/{listdir(data_path)[0]}")

df.head()

# Filter with the following columns
cols = ['SOC_CODE', 'TITLE', 'A_MEDIAN']

sub_df = df[cols]

# Retrieve OCC List
occ_df = pd.read_csv(grow_path)

occ_df.columns

# Keep SubFrame of just SOC_Code and GROW
sub_occ_df = occ_df[['SOC_Code', 'Grow']]
sub_occ_df.columns = [col.upper() for col in sub_occ_df.columns.tolist()]

sub_occ_df.head()
sub_df.shape

# Join with the wages dataframe
final_df = sub_df.merge(sub_occ_df, on=['SOC_CODE']).sort_values('TITLE')

# Check dtypes of dataframe
final_df.info()

# Cast type of A_MEDIAN as int32 in numpy
final_df['A_MEDIAN'] = final_df['A_MEDIAN'].astype(np.int32)

# Get the Wage Column by Dividing MEDIAN by 1000
final_df['WAGE'] = final_df['A_MEDIAN'] / 1000

# Drop the A_MEDIAN column
final_df.drop('A_MEDIAN', axis=1, inplace=True)

# Reset the index and drop extra index column
final_df = final_df.reset_index().drop('index', axis=1)

# Rearranging the index to start from 1
final_df.index = range(1, final_df.shape[0]+1)

# View table
final_df

# Save the file
final_df.to_csv(f"{data_path}/Wages_Table_Final.csv", index=False)

# Learnings
soc_codes = final_df['SOC_CODE'].unique()

df[df['SOC_CODE'].isin(soc_codes.tolist())]['TITLE'].unique()
```
